```python
import json
import pandas as pd
```

```python
tweets_data_path = '../Code/twitter_data.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```

```python
tweets = pd.DataFrame(columns=['text', 'hashtags'])

tweets['text'] = list(map(lambda tweet: tweet['text'], tweets_data))
for i in range(len(tweets_data)):
    tweets.iloc[i]['hashtags'] = list(map(lambda tweet: tweet['text'],tweets_data[i]['entities']['hashtags']))
    if len(tweets.iloc[i]['hashtags']) > 1:
        tweets.iloc[i]['hashtags'] = 'awaeseason4' if 'awaeseason4' in tweets.iloc[i]['hashtags'] else 'WarOnThanksgiving'
    elif len(tweets.iloc[i]['hashtags']) == 1:
        tweets.iloc[i]['hashtags'] = str(tweets.iloc[i]['hashtags'][0]) if tweets.iloc[i]['hashtags'][0] == 'awaeseason4' or tweets.iloc[i]['hashtags'][0] == 'WarOnThanksgiving' else None 
    else:
        tweets.iloc[i]['hashtags'] = None
```

```python
tweets.head()
```

```python
tweets = tweets.dropna().reset_index(drop=True)
tweets
```

```python
tweets.to_csv("tweets.csv",index=False)
```
